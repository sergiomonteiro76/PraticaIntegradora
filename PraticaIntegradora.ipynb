```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Criando um grafo não dirigido
grafo_nao_direcionado = nx.Graph()

# Adicionando 5 vértices ao grafo
num_vertices = 5
grafo_nao_direcionado.add_nodes_from(range(num_vertices))

# Adicionando 10 arestas com rótulos e pesos ao grafo não dirigido
num_arestas = 10
arestas_adicionadas = 0
while arestas_adicionadas < num_arestas:
    vertice_origem = random.randint(0, num_vertices - 1)
    vertice_destino = random.randint(0, num_vertices - 1)

    # Verificando se não é um laço e se a aresta já existe
    if vertice_origem != vertice_destino and not grafo_nao_direcionado.has_edge(vertice_origem, vertice_destino):
        peso = random.randint(1, 10)
        grafo_nao_direcionado.add_edge(vertice_origem, vertice_destino, weight=peso)
        grafo_nao_direcionado[vertice_origem][vertice_destino]['label'] = f'Peso: {peso}'
        arestas_adicionadas += 1

# Imprimindo o grafo não dirigido
print("Grafo Não Direcionado:")
print(grafo_nao_direcionado.edges(data=True))

# Desenhando o grafo não dirigido
plt.figure(figsize=(6, 6))
pos = nx.spring_layout(grafo_nao_direcionado)
nx.draw_networkx(grafo_nao_direcionado, pos, with_labels=True, node_size=500, node_color='skyblue', font_weight='bold')
labels = nx.get_edge_attributes(grafo_nao_direcionado, 'label')
nx.draw_networkx_edge_labels(grafo_nao_direcionado, pos, edge_labels=labels)
plt.title("Grafo Não Direcionado")
plt.axis('off')
plt.show()
```

```python
# Criando um grafo dirigido
grafo_dirigido = nx.DiGraph()

# Adicionando 5 vértices ao grafo
grafo_dirigido.add_nodes_from(range(num_vertices))

# Adicionando 10 arestas com rótulos e pesos ao grafo dirigido
arestas_adicionadas = 0
while arestas_adicionadas < num_arestas:
    vertice_origem = random.randint(0, num_vertices - 1)
    vertice_destino = random.randint(0, num_vertices - 1)

    # Verificando se não é um laço e se a aresta já existe
    if vertice_origem != vertice_destino and not grafo_dirigido.has_edge(vertice_origem, vertice_destino):
        peso = random.randint(1, 10)
        grafo_dirigido.add_edge(vertice_origem, vertice_destino, weight=peso)
        grafo_dirigido[vertice_origem][vertice_destino]['label'] = f'Peso: {peso}'
        arestas_adicionadas += 1

# Imprimindo o grafo dirigido
print("\nGrafo Direcionado:")
print(grafo_dirigido.edges(data=True))

# Desenhando o grafo dirigido
plt.figure(figsize=(6, 6))
pos = nx.spring_layout(grafo_dirigido)
nx.draw_networkx(grafo_dirigido, pos, with_labels=True, node_size=500, node_color='lightgreen', font_weight='bold')
labels = nx.get_edge_attributes(grafo_dirigido, 'label')
nx.draw_networkx_edge_labels(grafo_dirigido, pos, edge_labels=labels)
plt.title("Grafo Direcionado")
plt.axis('off')
plt.show()
```

# **3. Prática de Algoritmos de Grafos: Árvore Geradora Mínima e Caminho Mínimo**

## **Algoritmo de Kruskal**

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando um grafo com pesos para exemplificar o algoritmo de Kruskal
G_kruskal = nx.Graph()
G_kruskal.add_edge('A', 'B', weight=4)
G_kruskal.add_edge('A', 'C', weight=2)
G_kruskal.add_edge('B', 'C', weight=1)
G_kruskal.add_edge('B', 'D', weight=5)
G_kruskal.add_edge('C', 'D', weight=3)

# Visualizando o grafo original com os pesos das arestas
plt.figure(figsize=(4, 3))
pos = nx.spring_layout(G_kruskal)
edge_labels = nx.get_edge_attributes(G_kruskal, 'weight')
nx.draw(G_kruskal, pos, with_labels=True, font_weight='bold')
nx.draw_networkx_edge_labels(G_kruskal, pos, edge_labels=edge_labels)
plt.title('Grafo Original')
plt.show()

# Algoritmo de Kruskal para encontrar a Árvore Geradora Mínima
MST_kruskal = nx.minimum_spanning_tree(G_kruskal)

# Visualizando a Árvore Geradora Mínima encontrada pelo algoritmo de Kruskal
plt.figure(figsize=(4, 3))
pos = nx.spring_layout(MST_kruskal)
nx.draw(MST_kruskal, pos, with_labels=True, font_weight='bold')
edge_labels_mst = nx.get_edge_attributes(MST_kruskal, 'weight')
nx.draw_networkx_edge_labels(MST_kruskal, pos, edge_labels=edge_labels_mst)
plt.title('Árvore Geradora Mínima (Kruskal)')
plt.show()
```

## **Algoritmo de Prim**

```python
# Usando o algoritmo de Prim para encontrar a Árvore Geradora Mínima
MST_prim = nx.minimum_spanning_tree(G_kruskal, algorithm='prim')

# Visualizando a Árvore Geradora Mínima usando o algoritmo de Prim
plt.figure(figsize=(4, 3))
pos = nx.spring_layout(MST_prim)
nx.draw(MST_prim, pos, with_labels=True, font_weight='bold')
edge_labels_prim = nx.get_edge_attributes(MST_prim, 'weight')
nx.draw_networkx_edge_labels(MST_prim, pos, edge_labels=edge_labels_prim)
plt.title('Árvore Geradora Mínima (Prim)')
plt.show()
```

## **Algoritmo de Dijkstra**

```python
# Criando um grafo direcionado com pesos para exemplificar o algoritmo de Dijkstra
G_dijkstra = nx.DiGraph()
G_dijkstra.add_edge('A', 'B', weight=4)
G_dijkstra.add_edge('A', 'C', weight=2)
G_dijkstra.add_edge('B', 'C', weight=1)
G_dijkstra.add_edge('B', 'D', weight=5)
G_dijkstra.add_edge('C', 'D', weight=3)

# Usando o algoritmo de Dijkstra para encontrar o menor caminho de 'A' a 'D'
shortest_path = nx.shortest_path(G_dijkstra, 'A', 'D', weight='weight')

# Visualizando o menor caminho encontrado pelo algoritmo de Dijkstra
plt.figure(figsize=(4, 3))
pos = nx.spring_layout(G_dijkstra)
nx.draw(G_dijkstra, pos, with_labels=True, font_weight='bold')
path_edges = [(shortest_path[i], shortest_path[i+1]) for i in range(len(shortest_path)-1)]
nx.draw_networkx_edges(G_dijkstra, pos, edgelist=path_edges, edge_color='r', width=2)
edge_labels_dijkstra = nx.get_edge_attributes(G_dijkstra, 'weight')
nx.draw_networkx_edge_labels(G_dijkstra, pos, edge_labels=edge_labels_dijkstra)
plt.title('Menor Caminho (Dijkstra)')
plt.show()
```

## **Algoritmo A Estrela**

```python
import heapq
import networkx as nx
import matplotlib.pyplot as plt

def astar(graph, start, goal, heuristic):
    open_set = []
    closed_set = set()
    heapq.heappush(open_set, (0, start, [start]))

    while open_set:
        current_cost, current_node, path = heapq.heappop(open_set)

        if current_node == goal:
            return path

        if current_node in closed_set:
            continue

        closed_set.add(current_node)

        for neighbor, cost in graph[current_node].items():
            if neighbor in closed_set:
                continue

            heapq.heappush(open_set, (current_cost + cost['weight'] + heuristic(neighbor, goal), neighbor, path + [neighbor]))

    return None

# Exemplo de uso com pesos
graph = {
    'A': {'B': {'weight': 4}, 'C': {'weight': 2}},
    'B': {'A': {'weight': 4}, 'C': {'weight': 1}, 'D': {'weight': 5}},
    'C': {'A': {'weight': 2}, 'B': {'weight': 1}, 'D': {'weight': 3}},
    'D': {'B': {'weight': 5}, 'C': {'weight': 3}}
}

# Função heurística (distância em linha reta entre os nós)
def heuristic(node, goal):
    heuristic_costs = {
        'A': 4,
        'B': 3,
        'C': 2,
        'D': 0
    }
    return heuristic_costs[node]

start_node = 'A'
goal_node = 'D'

# Encontrando o caminho mais curto
shortest_path = astar(graph, start_node, goal_node, heuristic)
print(f"O menor caminho de {start_node} a {goal_node} é: {shortest_path}")

# Criando um grafo para visualização
G = nx.Graph()

# Adicionando nós e arestas com pesos ao grafo
for node, neighbors in graph.items():
    G.add_node(node)
    for neighbor, cost in neighbors.items():
        G.add_edge(node, neighbor, weight=cost['weight'])

# Adicionando o caminho mais curto ao grafo
edges = [(shortest_path[i], shortest_path[i+1]) for i in range(len(shortest_path)-1)]
G.add_edges_from(edges)

# Visualizando o grafo com o caminho mais curto e os pesos das arestas
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold')
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
nx.draw_networkx_edges(G, pos, edgelist=edges, edge_color='r', width=2)
plt.title(f"Caminho mais curto de {start_node} a {goal_node} com pesos das arestas")
plt.show()
```

# **5. Prática de Alocação Dinâmica de Memória: Listas Encadeadas**

## **Lista Simplesmente Encadeada: inserção e busca**


```python
class No:
    def __init__(self, valor):
        self.valor = valor
        self.proximo = None

class ListaSimplesmenteEncadeada:
    def __init__(self):
        self.cabeca = None

    def inserir(self, valor):
        novo_no = No(valor)
        if self.cabeca is None:
            self.cabeca = novo_no
        else:
            atual = self.cabeca
            while atual.proximo is not None:
                atual = atual.proximo
            atual.proximo = novo_no

    def buscar(self, valor):
        atual = self.cabeca
        while atual is not None:
            if atual.valor == valor:
                return True
            atual = atual.proximo
        return False

# Criando uma lista simplesmente encadeada e inserindo valores
lista = ListaSimplesmenteEncadeada()
lista.inserir(5)
lista.inserir(10)
lista.inserir(15)

# Realizando uma busca na lista
valor_busca = 10
resultado_busca = lista.buscar(valor_busca)
if resultado_busca:
    print(f"O valor {valor_busca} está na lista.")
else:
    print(f"O valor {valor_busca} não está na lista.")
```

```python
class Pilha:
    def __init__(self):
        self.items = []

    def vazia(self):
        return self.items == []

    def empilhar(self, item):
        self.items.append(item)

    def desempilhar(self):
        if not self.vazia():
            return self.items.pop()
        return None

    def topo(self):
        if not self.vazia():
            return self.items[-1]
        return None

# Testando a Pilha
pilha = Pilha()

# Empilhando alguns valores
pilha.empilhar(1)
pilha.empilhar(2)
pilha.empilhar(3)

# Imprimindo o topo
print("Topo da Pilha:", pilha.topo())

# Desempilhando e imprimindo os valores
print("Desempilhando:")
while not pilha.vazia():
    print(pilha.desempilhar())
```

## **Fila Circular**

```python
class FilaCircular:
    def __init__(self, tamanho):
        self.tamanho = tamanho
        self.items = [None] * tamanho
        self.inicio = 0
        self.tamanho_atual = 0

    def vazia(self):
        return self.tamanho_atual == 0

    def cheia(self):
        return self.tamanho_atual == self.tamanho

    def enfileirar(self, item):
        if not self.cheia():
            fim = (self.inicio + self.tamanho_atual) % self.tamanho
            self.items[fim] = item
            self.tamanho_atual += 1

    def desenfileirar(self):
        if not self.vazia():
            item = self.items[self.inicio]
            self.items[self.inicio] = None
            self.inicio = (self.inicio + 1) % self.tamanho
            self.tamanho_atual -= 1
            return item
        return None

    def frente(self):
        if not self.vazia():
            return self.items[self.inicio]
        return None

# Testando a Fila Circular
fila_circular = FilaCircular(5)

# Enfileirando alguns valores
fila_circular.enfileirar(1)
fila_circular.enfileirar(2)
fila_circular.enfileirar(3)

# Imprimindo a frente da fila circular
print("Frente da Fila Circular:", fila_circular.frente())

# Desenfileirando e imprimindo os valores
print("Desenfileirando:")
while not fila_circular.vazia():
    print(fila_circular.desenfileirar())
```

# **7. Prática de Métodos de Ordenação: básicos**


## **Bolha**

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        trocou = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                trocou = True
        if not trocou:
            break
    return arr

# Testando o Bubble Sort
lista_bolha = [64, 34, 25, 12, 22, 11, 90]
print("Lista original:", lista_bolha)
print("Ordenando com Bubble Sort:", bubble_sort(lista_bolha.copy()))
```

## **Seleção**

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        indice_minimo = i
        for j in range(i + 1, n):
            if arr[j] < arr[indice_minimo]:
                indice_minimo = j
        arr[i], arr[indice_minimo] = arr[indice_minimo], arr[i]
    return arr

# Testando o Selection Sort
lista_selecao = [64, 34, 25, 12, 22, 11, 90]
print("Lista original:", lista_selecao)
print("Ordenando com Selection Sort:", selection_sort(lista_selecao.copy()))
```

## **Inserção**


```python
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        chave = arr[i]
        j = i - 1
        while j >= 0 and chave < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = chave
    return arr

# Testando o Insertion Sort
lista_insercao = [64, 34, 25, 12, 22, 11, 90]
print("Lista original:", lista_insercao)
print("Ordenando com Insertion Sort:", insertion_sort(lista_insercao.copy()))
```

# **8. Prática de Métodos de Ordenação: avançados**

## **Quick Sort**

```python
def partition(arr, inicio, fim):
    pivo = arr[fim]
    i = inicio - 1

    for j in range(inicio, fim):
        if arr[j] <= pivo:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[fim] = arr[fim], arr[i + 1]
    return i + 1

def quick_sort(arr, inicio, fim):
    if inicio < fim:
        pivo = partition(arr, inicio, fim)
        quick_sort(arr, inicio, pivo - 1)
        quick_sort(arr, pivo + 1, fim)
    return arr

# Testando o Quick Sort
lista_quick = [64, 34, 25, 12, 22, 11, 90]
print("Lista original:", lista_quick)
print("Ordenando com Quick Sort:", quick_sort(lista_quick.copy(), 0, len(lista_quick) - 1))
```

## **Merge Sort**


```python
def merge_sort(arr):
    if len(arr) > 1:
        meio = len(arr) // 2
        esquerda = arr[:meio]
        direita = arr[meio:]

        merge_sort(esquerda)
        merge_sort(direita)

        i = j = k = 0

        while i < len(esquerda) and j < len(direita):
            if esquerda[i] < direita[j]:
                arr[k] = esquerda[i]
                i += 1
            else:
                arr[k] = direita[j]
                j += 1
            k += 1

        while i < len(esquerda):
            arr[k] = esquerda[i]
            i += 1
            k += 1

        while j < len(direita):
            arr[k] = direita[j]
            j += 1
            k += 1

    return arr

# Testando o Merge Sort
lista_merge = [64, 34, 25, 12, 22, 11, 90]
print("Lista original:", lista_merge)
print("Ordenando com Merge Sort:", merge_sort(lista_merge.copy()))
```

## **Heap Sort**

```python
def heapify(arr, n, i):
    maior = i
    esq = 2 * i + 1
    dir = 2 * i + 2

    if esq < n and arr[i] < arr[esq]:
        maior = esq

    if dir < n and arr[maior] < arr[dir]:
        maior = dir

    if maior != i:
        arr[i], arr[maior] = arr[maior], arr[i]
        heapify(arr, n, maior)

def heap_sort(arr):
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

    return arr

# Testando o Heap Sort
lista_heap = [64, 34, 25, 12, 22, 11, 90]
print("Lista original:", lista_heap)
print("Ordenando com Heap Sort:", heap_sort(lista_heap.copy()))
```

# **9. Prática de Métodos de Pesquisa: estruturas sequenciais**


## **Busca Sequencial**

```python
def busca_sequencial(lista, item):
    for i in range(len(lista)):
        if lista[i] == item:
            return i
    return -1

# Testando a Busca Sequencial
lista_sequencial = [3, 7, 2, 10, 5]
elemento = 10
print("Lista:", lista_sequencial)
print(f"Buscando o elemento {elemento} na posição:", busca_sequencial(lista_sequencial, elemento))
```

## **Busca Binária**


```python
def busca_binaria(lista, item):
    inicio = 0
    fim = len(lista) - 1
    while inicio <= fim:
        meio = (inicio + fim) // 2
        if lista[meio] == item:
            return meio
        elif lista[meio] < item:
            inicio = meio + 1
        else:
            fim = meio - 1
    return -1

# Testando a Busca Binária
lista_ordenada = [2, 3, 5, 7, 10]
elemento = 7
print("Lista ordenada:", lista_ordenada)
print(f"Buscando o elemento {elemento} na posição:", busca_binaria(lista_ordenada, elemento))
```

## **Árvore Binária Balanceada**


```python
class No:
    def __init__(self, valor):
        self.valor = valor
        self.esquerda = None
        self.direita = None
        self.altura = 1

class ArvoreAVL:
    def altura(self, no):
        if not no:
            return 0
        return no.altura

    def rotacao_simples_direita(self, z):
        y = z.esquerda
        T3 = y.direita

        # Realiza a rotação
        y.direita = z
        z.esquerda = T3

        # Atualiza alturas
        z.altura = 1 + max(self.altura(z.esquerda), self.altura(z.direita))
        y.altura = 1 + max(self.altura(y.esquerda), self.altura(y.direita))

        return y

    def rotacao_simples_esquerda(self, z):
        y = z.direita
        T2 = y.esquerda

        # Realiza a rotação
        y.esquerda = z
        z.direita = T2

        # Atualiza alturas
        z.altura = 1 + max(self.altura(z.esquerda), self.altura(z.direita))
        y.altura = 1 + max(self.altura(y.esquerda), self.altura(y.direita))

        return y

    def inserir(self, no, valor):
        if not no:
            return No(valor)

        if valor < no.valor:
            no.esquerda = self.inserir(no.esquerda, valor)
        else:
            no.direita = self.inserir(no.direita, valor)

        no.altura = 1 + max(self.altura(no.esquerda), self.altura(no.direita))

        balanceamento = self.altura(no.esquerda) - self.altura(no.direita)

        # Rotações
        # Rotação Simples à Esquerda
        if balanceamento > 1 and valor < no.esquerda.valor:
            return self.rotacao_simples_direita(no)

        return no

    def pre_ordem(self, no):
        if not no:
            return []

        resultado = [no.valor]
        resultado += self.pre_ordem(no.esquerda)
        resultado += self.pre_ordem(no.direita)

        return resultado

    def imprime_arvore(self, no, espaco):
        if not no:
            return

        espaco += 10

        if no.direita:
            self.imprime_arvore(no.direita, espaco)

        print()
        for i in range(10, espaco):
            print(" ", end="")

        print(no.valor)

        if no.esquerda:
            self.imprime_arvore(no.esquerda, espaco)

# Função para inserir valores na árvore AVL
def inserir_valores(arvore, valores):
    raiz = None
    for valor in valores:
        raiz = arvore.inserir(raiz, valor)
    return raiz

# Criando a árvore AVL
arvore = ArvoreAVL()
valores = [10, 20, 30, 40, 50, 25]

# Inserindo valores na árvore AVL
raiz = inserir_valores(arvore, valores)

# Imprimindo a árvore AVL original em pré-ordem
print("Árvore AVL Original (Pré-ordem):")
print(arvore.pre_ordem(raiz))

# Realizando uma rotação simples à esquerda
raiz = arvore.rotacao_simples_esquerda(raiz)
print("\nÁrvore AVL após Rotação Simples à Esquerda (Pré-ordem):")
print(arvore.pre_ordem(raiz))
```

```python
import pydot
from IPython.display import Image
# Definindo as informações do autômato
states = {"q0", "q1", "q2"}
input_symbols = {"0", "1"}
transitions = {
    "q0": {"0": {"q2"}, "1": {"q1"}},
    "q1": {"0": {"q0", "q2"}, "1": {"q2"}},
    "q2": {},
}
initial_state = "q0"
final_states = {"q0"}

# Criando o gráfico
graph = pydot.Dot(graph_type='digraph')

# Adicionando os estados
for state in states:
    if state in final_states:
        graph.add_node(pydot.Node(state, shape='doublecircle'))
    else:
        graph.add_node(pydot.Node(state, shape='circle'))

# Adicionando as transições
for from_state, transitions in transitions.items():
    for symbol, to_states in transitions.items():
        for to_state in to_states:
            graph.add_edge(pydot.Edge(from_state, to_state, label=symbol))

# Salvar o gráfico como um arquivo PNG
graph.write_png('/content/nfa1.png')

# Exibir a imagem no Colab
Image('/content/nfa1.png')
```

## **AFN com Transiçoes Vazias**

```python
import pydot
from IPython.display import Image

# Definindo as informações do autômato com transições vazias
states = {"q0", "q1", "q2", "q3"}
input_symbols = {"0", "1"}

# Transições com símbolos de entrada
transitions = {
    "q0": {"0": {"q1"}, "1": {"q2"}},
    "q1": {"": {"q3"}},
    "q2": {"0": {"q2"}, "1": {"q3"}},
    "q3": {"": {"q0"}},
}

# Criando o gráfico
graph = pydot.Dot(graph_type='digraph')

# Adicionando os estados
for state in states:
    graph.add_node(pydot.Node(state, shape='circle'))

# Adicionando as transições com rótulos
for from_state, transitions in transitions.items():
    for symbol, to_states in transitions.items():
        for to_state in to_states:
            if symbol:  # Se não for uma transição vazia, adicione o rótulo
                edge_label = f"{symbol}"
            else:  # Se for uma transição vazia, o rótulo será vazio
                edge_label = "ε"

            graph.add_edge(pydot.Edge(from_state, to_state, label=edge_label))

# Salvar o gráfico como um arquivo PNG
graph.write_png('/content/nfa_epsilon.png')

# Exibir a imagem no Colab
Image('/content/nfa_epsilon.png')
# Exibir a imagem no Colab
Image('/content/nfa_epsilon.png')
```

## **Equivalências Entre os Autômatos**

```python
def epsilon_closure(states, transitions):
    epsilon_states = set(states)
    queue = list(states)

    while queue:
        state = queue.pop(0)
        epsilon_transitions = transitions.get((state, ''), set())

        for epsilon_state in epsilon_transitions:
            if epsilon_state not in epsilon_states:
                epsilon_states.add(epsilon_state)
                queue.append(epsilon_state)

    return epsilon_states

def afn_to_afd(states, input_symbols, transitions):
    dfa_states = [epsilon_closure({'q0'}, transitions)]
    dfa_transitions = {}
    queue = list([dfa_states[0]])

    while queue:
        current_states = queue.pop(0)

        for symbol in input_symbols:
            next_states = set()
            for state in current_states:
                next_states |= transitions.get((state, symbol), set())

            epsilon_states = epsilon_closure(next_states, transitions)

            if epsilon_states not in dfa_states:
                dfa_states.append(epsilon_states)
                queue.append(epsilon_states)

            dfa_transitions[tuple(current_states), symbol] = tuple(epsilon_states)

    return dfa_states, dfa_transitions

# Definição do AFN
states_afn = {"q0", "q1", "q2", "q3"}
input_symbols_afn = {"0", "1"}

transitions_afn = {
    ("q0", ""): {"q1", "q2"},
    ("q1", "0"): {"q3"},
    ("q2", "1"): {"q3"},
    ("q3", ""): {"q0"}
}

# Conversão do AFN para AFD
dfa_states, dfa_transitions = afn_to_afd(states_afn, input_symbols_afn, transitions_afn)

# Exibição do AFD resultante
print("Estados do AFD:", dfa_states)
print("Transições do AFD:", dfa_transitions)
```

```python
#Verificar se um endereço de e-mail é válido
import re

def is_valid_email(email):
    email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    return re.match(email_pattern, email) is not None

# Testando endereços de e-mail
emails = ["email@example.com", "invalid_email@", "another_email.com"]

for email in emails:
    if is_valid_email(email):
        print(f"{email} é um endereço de e-mail válido.")
    else:
        print(f"{email} NÃO é um endereço de e-mail válido.")
```

```python
#Validar senhas fortes
import re

def is_strong_password(password):
    # Pelo menos 8 caracteres, incluindo uma letra maiúscula, uma minúscula, um número e um caractere especial
    password_pattern = r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*?&])[A-Za-z\d@$!%*?&]{8,}$'
    return re.match(password_pattern, password) is not None

# Testando senhas
passwords = ["SenhaForte@123", "senhafraca", "A1b@", "Senha123"]

for password in passwords:
    if is_strong_password(password):
        print(f"{password} é uma senha forte.")
    else:
        print(f"{password} NÃO é uma senha forte.")
```

## **Correspondência e Substituição de Texto**

```python
#Encontrar palavras começando com uma determinada letra
import re

def find_words_starting_with_letter(text, letter):
    pattern = fr'\b{letter}\w+\b'
    return re.findall(pattern, text, re.IGNORECASE)

# Texto de exemplo
text = "O rato roeu a roupa do rei de Roma."

# Encontrar palavras começando com 'r' (maiúsculas ou minúsculas)
letter = 'r'
matches = find_words_starting_with_letter(text, letter)
print(f"Palavras começando com '{letter}': {matches}")
```

```python
# Substituir todas as ocorrências de uma palavra por outra
import re

def replace_word(text, old_word, new_word):
    pattern = fr'\b{old_word}\b'
    return re.sub(pattern, new_word, text)

# Texto de exemplo
text = "O rato roeu a roupa do rei de Roma."

# Substituir todas as ocorrências de 'rato' por 'gato'
new_text = replace_word(text, 'rato', 'gato')
print("Texto original:", text)
print("Texto modificado:", new_text)
```

```python
#Remover números de telefone de um texto
import re

def remove_phone_numbers(text):
    pattern = r'\d{3}-\d{3}-\d{4}'
    return re.sub(pattern, '', text)

# Texto de exemplo
text = "Entre em contato no número 123-456-7890. Obrigado!"

# Remover números de telefone do texto
new_text = remove_phone_numbers(text)
print("Texto original:", text)
print("Texto sem números de telefone:", new_text)
```

## **Grupos de Captura**

```python
# Separar o nome e sobrenome de uma lista de nomes
import re

def extract_names(full_names):
    pattern = r'(\w+)\s(\w+)'
    matches = [re.match(pattern, name) for name in full_names]
    return [(match.group(1), match.group(2)) for match in matches if match]

# Lista de nomes completos
names = ["John Doe", "Alice Cooper", "Bob Marley", "Eva Mendes"]

# Extrair nome e sobrenome
extracted_names = extract_names(names)
print("Nomes e sobrenomes extraídos:")
for first_name, last_name in extracted_names:
    print(f"Nome: {first_name}, Sobrenome: {last_name}")
```

```python
 #Extrair números de telefone e seus códigos de área
import re

def extract_phone_numbers(text):
    pattern = r'(\d{3})-(\d{3}-\d{4})'
    matches = re.findall(pattern, text)
    return matches

# Texto com números de telefone
text = "Números de telefone: 123-456-7890, 987-654-3210, 111-222-3333"

# Extrair números de telefone e códigos de área
phone_numbers = extract_phone_numbers(text)
print("Números de telefone e códigos de área:")
for area_code, phone_number in phone_numbers:
    print(f"Código de Área: {area_code}, Número: {phone_number}")
```

```python
#Extrair tags HTML e seus conteúdos
import re

def extract_html_tags(html):
    pattern = r'<(\w+)>.*?</\1>'
    matches = re.findall(pattern, html)
    return matches

# Texto com tags HTML
html_text = "<p>Este é um <b>texto</b> de exemplo.</p><div>Conteúdo da <b>div</b>.</div>"

# Extrair tags HTML e seu conteúdo
html_tags = extract_html_tags(html_text)
print("Tags HTML e seus conteúdos:")
for tag in html_tags:
    print(f"Tag: {tag}")
```

## **Biblioteca joblib**

```python
from joblib import Parallel, delayed

def factorial(number):
    result = 1
    for i in range(1, number + 1):
        result *= i
    return result

if __name__ == "__main__":
    numbers = [3, 4, 5, 6, 7]

    results = Parallel(n_jobs=-1)(delayed(factorial)(num) for num in numbers)

    print("Resultado usando joblib:", results)
```

## **Paradigma MapReduce**

```python
#Exemplo 01
from functools import reduce

# Dados de exemplo
numbers = [1, 2, 3, 4, 5]

# Map: Multiplicar cada número por ele mesmo
mapped = map(lambda x: x * x, numbers)

# Reduce: Somar todos os números mapeados
result = reduce(lambda x, y: x + y, mapped)

print("Resultado usando MapReduce:", result)
```

```python
#Exemplo 02
from functools import reduce

# Função Map
def mapper(number):
    return number ** 2

# Função Reduce
def reducer(acc, value):
    return acc + value

if __name__ == "__main__":
    numbers = [1, 2, 3, 4, 5]

    # Map
    mapped_values = map(mapper, numbers)

    # Reduce
    result = reduce(reducer, mapped_values)

    print("Resultado usando MapReduce:", result)
```

```python
from flask import Flask, jsonify, request

app = Flask(__name__)

# Simulando uma "base de dados" em memória
users = [
    {"id": 1, "name": "Alice"},
    {"id": 2, "name": "Maria"},
    {"id": 3, "name": "Wendy"}
]


# Rota para obter todos os usuários
@app.route('/users', methods=['GET'])
def get_users():
    return jsonify(users)


# Rota para obter um usuário por ID
@app.route('/users/<int:user_id>', methods=['GET'])
def get_user(user_id):
    user = next((user for user in users if user["id"] == user_id), None)
    if user:
        return jsonify(user)
    return jsonify({"message": "User not found"}), 404


# Rota para adicionar um novo usuário
@app.route('/users', methods=['POST'])
def add_user():
    new_user = request.json
    if not new_user or 'name' not in new_user:
        return jsonify({"message": "Invalid data"}), 400

    # Gerando um ID para o novo usuário (apenas para exemplo, não é a melhor prática)
    new_user_id = max(user["id"] for user in users) + 1
    new_user["id"] = new_user_id

    users.append(new_user)
    return jsonify({"mensagem": "Usuário adicionado com sucesso", "user": new_user}), 201


if __name__ == '__main__':
    app.run(debug=True)

```

```python
from flask import Flask, jsonify, request

app = Flask(__name__)

# Rota para adição
@app.route('/somar', methods=['POST'])
def somar():
    data = request.get_json()
    num1 = data.get('num1', 0)
    num2 = data.get('num2', 0)
    result = num1 + num2
    return jsonify({'result': result})

# Rota para divisão
@app.route('/dividir', methods=['POST'])
def dividir():
    data = request.get_json()
    num1 = data.get('num1', 0)
    num2 = data.get('num2', 0)
    if num2 == 0:
        return jsonify({'error': 'Divisão por zero não é permitida'})
    result = num1 / num2
    return jsonify({'result': result})

if __name__ == '__main__':
    app.run(debug=True)
```

**Obter o preço médio de uma determinada ação nos últimos 30 dias**

```python
import yfinance as yf
from flask import Flask, jsonify

app = Flask(__name__)

class StockService:
    def get_avg_close_price(self, symbol):
        try:
            stock = yf.Ticker(symbol)
            hist = stock.history(period='1mo')  # Obtém o histórico dos últimos 30 dias
            if not hist.empty:
                avg_close_price = hist['Close'].mean()
                return avg_close_price
            else:
                return None
        except Exception as e:
            print(f"Erro ao obter informações da ação {symbol}: {e}")
            return None

@app.route('/stock/<symbol>/avgClosePrice', methods=['GET'])
def get_average_close_price(symbol):
    stock_service = StockService()
    avg_close_price = stock_service.get_avg_close_price(symbol)

    if avg_close_price is not None:
        return jsonify({'averageClosePrice': avg_close_price})
    else:
        return jsonify({'error': f"Não foi possível obter informações sobre {symbol} nos últimos 30 dias"}), 404

if __name__ == '__main__':
    app.run(debug=True)
```

## **Serviço de Cotação**

```python
import yfinance as yf
from flask import Flask, jsonify

app = Flask(__name__)

class StockService:
    def get_close_prices_last_5_days(self, symbol):
        try:
            stock = yf.Ticker(symbol)
            hist = stock.history(period='5d')  # Obtém o histórico dos últimos 10 dias
            if not hist.empty:
                close_prices = hist['Close'].tolist()
                return close_prices
            else:
                return None
        except Exception as e:
            print(f"Erro ao obter informações da ação {symbol}: {e}")
            return None

@app.route('/stock/<symbol>/closePricesLast5Days', methods=['GET'])
def get_close_prices_last_5_days(symbol):
    stock_service = StockService()
    close_prices = stock_service.get_close_prices_last_5_days(symbol)

    if close_prices is not None:
        return jsonify({'closePricesLast5Days': close_prices})
    else:
        return jsonify({'error': f"Não foi possível obter informações sobre os preços de fechamento de {symbol} nos últimos 5 dias"}), 404

if __name__ == '__main__':
    app.run(debug=True)
```
